```python
import gzip
import json

file_path = '/abcd_v1.1.json.gz'

def read_json(json_data):
    print(json.dumps(json_data, indent=4))

with gzip.open(file_path, 'rt') as file:
    data = json.load(file)

training_input_output = []
test_input_output = []

training_convos = data.get('train')

test_convos = data.get('test')

for convo in training_convos:
  inpt = convo.get('original')
  outpt = convo.get('scenario')
  training_input_output.append([inpt, outpt])

for convo in test_convos:
  inpt = convo.get('original')
  outpt = convo.get('scenario')
  test_input_output.append([inpt, outpt])

with open('training_input_output.json', 'w') as train_file:
    json.dump(training_input_output, train_file, indent=4)

with open('test_input_output.json', 'w') as test_file:
    json.dump(test_input_output, test_file, indent=4)

print(json.dumps(training_input_output[0], indent=4))

```
